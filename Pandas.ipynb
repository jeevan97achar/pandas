## Loading data into Pandas

```python
import pandas as pd
```

```python
# Loading the dataset

df = pd.read_csv('pokemon_dataset.csv')
print(df)
```

Different file formats can be loaded like,
* pd.read_xlsx('*fileName*')
* pd.read_csv('*fileName*', delimeter='*delimeter*')

```python
# To see just the top 5 (default number) rows

# You can pass an integer inside the method to vary the number of attributes
df.head()
```

```python
# To see just the bottom 5 (default number) rows

df.tail()
```

## Reading Data in Pandas

```python
# Read Headers
df.columns
```

```python
# Read each column

df['Name']
```

```python
# Alternative
df.Name
```

```python
# To fetch attributes based on indexing
df['Name'][0:5]
```

```python
# Passing a list to select multiple attributes
df[['Name', 'Type 1', 'Speed']]
```

```python
# Printing each row

df.iloc[1]
```

```python
# To query multiple rows

df.iloc[1:4]
```

```python
# Read a specific Location iloc(R,C)

df.iloc[3,1]
```
