## Making Changes to the Data

```python
# Adding a new column
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head()
```

```python
# Deleting a column

df = df.drop(columns=['Total'])
df.head()
```

```python
df['Total'] = df.iloc[:,4:10].sum(axis=1)
df.head()
```

```python
# Moving the column positions

cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]]+cols[4:12]]
df.head()
```

## Saving our dataframe

```python
df.to_csv('Modified.csv', index=False)
df.to_excel('Modified.xlsx', index=False)
df.to_csv('Modified.txt', index=False, sep='\t')
```

## Filtering Data

```python
new_df = df.loc[(df['Type 1']=='Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df
```

```python
# To reset the index in the new dataset
new_df = new_df.reset_index()
new_df

# To drop the original index
new_df = new_df.reset_index(drop=True)
new_df
```

```python
# To filter based on the string values of the attributes

df.loc[df['Name'].str.contains('Mega')]

# To get the inverse
df.loc[~df['Name'].str.contains('Mega')]
```

```python
# To filter using regex

import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]

# Here above flags=re.I ignores the case sensitivity
```

```python
# To get all the names starting from the letters 'pi'

df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]

# Here in the regex condition "^" is used to say that the name should begin with 'pi'
# "*" symbol means to say zero or more
```

## Conditional Changes

```python
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
# df.loc[df['Type 1'] == 'Fire']
df

# In the above code 
# df.loc[<where to make the change>, <label of the column to be changed>] = '<New variable>'

# Lists can be passed to change multiple columns at once
```

## Aggregate Statistics (Groupby)

```python
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
df['Count'] = 1
```

```python
df.groupby(['Type 1', 'Type 2']).count()['Count']
```

## Working with large amounts of data

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.csv', chunksize=5):
  results = df.groupby(['Type 1']).count()

  new_df = pd.concat([new_df, results])
```
